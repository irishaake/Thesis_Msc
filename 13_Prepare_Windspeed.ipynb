## Prepare Wind Speed

In this notebook, monthly averaged ERA5-Land wind speed data from 1981 - 2025 is processed. 
The following steps are included:
- Combine u and v component of wind to total wind speed
- Crop data to Germany
- Crop to forest area
- Save datasets for needleleaved and broadleaved forests

```python
# import libraries
import os
import xarray as xr
import rioxarray as rioxr
import geopandas as gpd
import odc.stac
import rasterio
import warnings
```

```python
warnings.filterwarnings("ignore", category = UserWarning)
warnings.filterwarnings("ignore", category = RuntimeWarning)
```

```python
# open dataset
era5 = xr.open_dataset("./Raw/ERA5-Land/data_1.nc", decode_coords="all",decode_times=True)
```

```python
era5
```

```python
# rename dimensions to fit other data
era5 = era5.rename({"valid_time":"time",
                            "longitude": "lon",
                            "latitude": "lat"})
```

```python
# create new variable that combines u and v component of wind
# source: https://knowledge.dea.ga.gov.au/notebooks/How_to_guides/External_data_ERA5_Climate/
era5["wind10m"] = (era5["u10"]**2 + era5["v10"]**2)**0.5
```

```python
# extract wind data
wind = era5.drop_vars(["number", "expver", "u10", "v10", "t2m"])
wind
```

```python
# set unit in variable attributes
wind.wind10m.attrs["units"] = "m/s"
```

```python
# plot exemplary data of one month to get a first impression
wind.wind10m.sel(time='2024-06-01').plot()
```

```python
# check crs
wind.rio.crs
```

```python
# assign crs to data
wind = wind.rio.write_crs('WGS 1984')
```

```python
# load Germany shapefile 
ger = gpd.read_file("./GER.shp")
```

```python
# check if crs matches data
ger.crs
```

```python
# extract one date to create mask of germany with dimensions of climate data
wind_ex = wind.sel(time = "1981-08-01", method = "nearest")
```

```python
# create mask of research area using the dimensions of the exemplary climate data
ger_mask = rasterio.features.geometry_mask(ger.geometry, 
                                            out_shape=wind_ex.odc.geobox.shape,
                                            transform=wind_ex.odc.geobox.affine,
                                            all_touched=False,
                                            invert=False)
```

```python
ger_mask = xr.DataArray(ger_mask, dims=("lat", "lon"))
```

```python
# create loop to process wind speed data per year
for year in range(1981, 2025):
    
    # select all monthly data within this year
    wind_yearly = wind.sel(time = wind.time.dt.year.isin(year))

    # apply mask of research area to dataset
    wind_ger = wind_yearly["wind10m"].where(~ger_mask)

    # load forest area from ESA Landcover data for corresponding year
    # attention: forest data only available for 1992 - 2022 -> choose 1992 for climate data of 1988-1992
    # and 2022 for climate data of 2022-2024
    if year <= 1992:
        forest_year = 1992
    elif year >= 2022:
        forest_year = 2022
    else: forest_year = year

    forest = xr.open_dataset(f"./Forest_GER/Forest_GER_{forest_year}.nc", 
                                decode_coords = "all", 
                                decode_times= True)
    
    # set crs for climate data 
    wind_ger.rio.write_crs(4326, inplace = True)
    
    # set spatial dims - otherwise resampling raises an error
    wind_ger.rio.set_spatial_dims(x_dim="lon", y_dim="lat", inplace = True)

    # resample climate dataset to match resolution of forest dataset
    wind_match = wind_ger.rio.reproject_match(forest)

    # rename dimensions - they got renamed while resampling
    wind_match = wind_match.rename({'y':'lat', 'x':'lon'})

    # mask with forest data
    # broadleaved forest:
    wind_BL = wind_match.where(forest.lccs_class.sel(time=f"{forest_year}-01-01") == 60, 
                                drop = False)

    # needleleaved forest:
    NL_mask = forest.lccs_class.sel(time=f"{forest_year}-01-01").isin([70, 80]).compute()
    wind_NL = wind_match.where(NL_mask, drop = False)

    # compress data to reduce size
    comp = dict(zlib=True, complevel=4)
    wind_BL.encoding.update(comp)
    wind_NL.encoding.update(comp)

    # save as netCDF4 datasets
    wind_BL.to_netcdf(f"./Climate/Windspeed/Monthly/wind10m_mon_mean_{year}_BL.nc")
    wind_NL.to_netcdf(f"./Climate/Windspeed/Monthly/wind10m_mon_mean_{year}_NL.nc")
```
